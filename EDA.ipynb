```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
insurance = pd.read_csv('insurance(1).csv')
```

```python
insurance.head()
```

```python
insurance.columns=insurance.columns.str.replace(' ','')
insurance.describe()
```

```python
insurance.shape
```

```python
insurance.info()
```

```python
print(insurance["smoker"].value_counts())
```

```python
print(insurance["sex"].value_counts())
```

```python
print(insurance["age"].max())
print(insurance["children"].max())
print(insurance["charges"].max())
print(insurance["bmi"].max())
```

```python
insurance.groupby(['sex','smoker']).mean()
```

```python
print(insurance.isnull().sum())
```

```python
plt.figure(figsize=(50,40))
insurance.hist()
plt.suptitle("histogram", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=insurance, x='sex',y ='charges')
plt.title("Distribution of charges by sex")
plt.show()
          
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=insurance, x='smoker',y ='charges')
plt.title("Distribution of charges by smokes")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=insurance, x='bmi',y ='charges')
plt.title("Distribution of charges by bmi")
plt.show()
```

```python
print(insurance.groupby(by="sex").mean())
insurance.groupby(by="sex").mean().plot(kind="bar")
plt.title("Sex vs Mean")
plt.ylabel("Means")
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))

insurance.groupby(by="smoker").mean()
insurance.groupby(by="smoker").mean().plot(kind="bar")
plt.title("smoker vs mean")
plt.ylabel("mean")
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1,1))
```

```python
plt.figure(figsize=(8,4))
sns.heatmap(insurance.corr(),cmap='Greens', annot=False)
```
